```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 3. PLOT PIE CHART

```python
values = [20, 55, 5, 17, 3]
colors = ['g', 'r', 'y', 'b', 'm']
labels = ['AAPL', 'GOOGL', 'T', 'TSLA', 'AMZN']
explode = [0, 0, 0, 0, 0.2]
# Use matplotlib to plot a pie chart 

plt.figure(figsize = (10, 10))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Stock Portfolio')
```

# 4. PLOT HISTOGRAMS

```python
# A histogram represents data using bars of various heights. 
# Each bar groups numbers into specific ranges. 
# Taller bars show that more data falls within that specific range (higher frequency)
mu = daily_return_df['FB'].mean()
sigma = daily_return_df['FB'].std()

num_bins = 40
plt.figure(figsize = (15, 10))
plt.hist(daily_return_df['FB'], num_bins, facecolor = 'blue'); #semicolon gets rid of extra text

plt.grid()
plt.title(f'Histogram: mu = {mu}, sigma = {sigma}');
```

The histogram for TWITTER returns using 30 bins


```python
mu = daily_return_df['TWTR'].mean()
sigma = daily_return_df['TWTR'].std()

num_bins = 40
plt.figure(figsize = (15, 10))
plt.hist(daily_return_df['TWTR'], num_bins, facecolor = 'lightblue'); 

plt.title(f'Histogram: mu = {mu}, sigma = {sigma}');
```

# 8. SEABRON SCATTERPLOT & COUNTPLOT

```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data drom the Sklearn library
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
# Check out the head of the dataframe
df_cancer.head(10)
```

```python
# Plot scatter plot between mean area and mean smoothness
plt.figure(figsize = (10,10))

sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue='target', data=df_cancer)
```

```python
# Print out countplot to know how many samples belong to class #0 and #1
plt.figure(figsize = (10,10))
sns.countplot(x='target', data=df_cancer, label='Count')
```

The scatterplot between the mean radius and mean area


```python
sns.scatterplot(x='mean radius', y='mean area', hue='target', data = df_cancer)
```

# 9. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS


```python
# Plot the pairplot
sns.pairplot(df_cancer, hue = 'target', 
             vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])

```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize = (30, 30)) 
sns.heatmap(df_cancer.corr(), annot = True)
```

```python
# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 
sns.distplot(df_cancer['mean radius'], bins = 25, color = 'b')
```

Two separate distplot for each target class #0 and target class #1

```python
class_0_df = df_cancer[ df_cancer['target'] == 0];
class_1_df = df_cancer[ df_cancer['target'] == 1];

class_0_df

class_1_df

# Plot the distplot for both classes
plt.figure(figsize=(10, 7));
sns.distplot(class_0_df['mean radius'], bins = 25, color = 'blue');
sns.distplot(class_1_df['mean radius'], bins = 25, color = 'red');
plt.grid();
```
